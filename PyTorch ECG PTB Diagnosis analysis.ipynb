```python
normal_df['label'] = np.array([0 for i in range(len(normal_df))])
abnormal_df['label'] = np.array([1 for i in range(len(abnormal_df))])
```

```python
from sklearn.model_selection import train_test_split
train_df,test_df = train_test_split(df,test_size=0.33)
```

```python
class ECG_Dataset(Dataset):
    def __init__(self,df):
        self.df = df
        self.features = df.drop('label',axis=1).values
        self.labels = df['label'].values
    def __len__(self):
        return len(self.df)
    def __getitem__(self,idx):
        feature = torch.Tensor(self.features[idx])
        label = torch.tensor(self.labels[idx])
        return feature,label
```

```python
train_dataset = ECG_Dataset(train_df)
test_dataset = ECG_Dataset(test_df)
```

```python
def pad_collate(batch):
    data = [item[0] for item in batch]
    labels = [item[1] for item in batch]
    max_size = max([item.size(0) for item in data])
    padded_data = [F.pad(item, (0, max_size - item.size(0))) for item in data]
    return torch.stack(padded_data), torch.tensor(labels)
```

```python
train_batch_size = 150
test_batch_size = 150
train_loader = DataLoader(train_dataset,batch_size=train_batch_size,shuffle=True,collate_fn=pad_collate)
test_loader = DataLoader(test_dataset,batch_size=test_batch_size,shuffle=False,collate_fn=pad_collate)
```

```python
class LSTM(nn.Module):
    
    def __init__(self,in_size,hidden_size,out_size=2,num_layers=1,bidirectional=False,p=0.4):
        super(LSTM,self).__init__()
        self.num_layers = num_layers
        self.num_dir = 2 if bidirectional else 1
        self.hidden_size = hidden_size
        self.lstm = nn.LSTM(in_size,hidden_size,num_layers=num_layers,bidirectional=bidirectional,batch_first=True)
        self.fc = nn.Linear(self.num_dir * self.hidden_size,out_size)
        self.dropout = nn.Dropout(p)
    def forward(self,x,hidden):
        lstm_out,hidden = self.lstm(x,hidden)
        lstm_out = self.dropout(lstm_out)
        pred = self.fc(lstm_out[:,-1,:])
        return pred
    def init_hidden(self,batch_size):
        hidden = (torch.zeros(self.num_dir*self.num_layers,batch_size,self.hidden_size).to(device),
                 torch.zeros(self.num_dir*self.num_layers,batch_size,self.hidden_size).to(device))
        return hidden
```

```python
model = LSTM(in_size=188,hidden_size=32,out_size=2,num_layers=1,bidirectional=False,p=0.4).to(device)
```

```python
num_epochs = 3
optimizer = torch.optim.Adam(model.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
```

```python
for epoch in range(num_epochs):
    model.train()
    for X_train,y_train in train_loader:
        X_train,y_train = X_train.to(device),y_train.to(device)
        X_train = X_train.unsqueeze(1)
        hidden = model.init_hidden(X_train.size(0))
        y_pred = model.forward(X_train,hidden)
        loss = criterion(y_pred,y_train)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f"epoch : {epoch+1} loss : {loss.item()}")
```

```python
from sklearn.metrics import classification_report

all_preds = []
all_labels = []

model.eval()
with torch.no_grad():
    for X_test, y_test in test_loader:
        X_test, y_test = X_test.to(device), y_test.to(device)
        hidden = model.init_hidden(X_test.size(0))
        
        # Add channel dimension if needed
        X_test = X_test.unsqueeze(1) 
        
        # Forward pass
        pred = model(X_test, hidden)
        pred = torch.argmax(pred, dim=1)
        
        # Collect predictions and labels
        all_preds.extend(pred.cpu().numpy())
        all_labels.extend(y_test.cpu().numpy())

# Use classification_report with the collected predictions and labels
print(classification_report(all_labels, all_preds))
```
